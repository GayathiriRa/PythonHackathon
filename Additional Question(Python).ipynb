```python
#Q:3  Finding union, intersect,differnce and Symmerterical difference of two arrays

a =ny.array([1,2,3,4,5,6])
print(a)

b =ny.array([5,6,7,8,9,10])
print(b)

c = a.reshape(3,2)
print(c)


d = b.reshape(3,2)
print(d)


abunion = ny.union1d(a,b)

abintersect = ny.intersect1d(a, b, assume_unique=True)

ab_difference = ny.setdiff1d(a, b, assume_unique=True)
ba_difference = ny.setdiff1d(b, a, assume_unique=True)
absymdiff = ny.setxor1d(a, b, assume_unique=True)



print("Union of a and b", abunion)
print("intersect of a and b",abintersect )
print("difference of a and b",ab_difference )
print("difference of b and b",ba_difference)
print("Symmetricdifference of a and b",absymdiff)
```

```python
#Q6: convert FirstName into Upper and Lower cases in DataFrame
Patient_data = pd.read_excel('HospitalDatabase.xlsx',sheet_name='Patients',usecols='B')

df = pd.DataFrame(Patient_data)
df1 = pd.DataFrame(Patient_data)
print("Upper cases in First Name")
df['FirstName'] = list(map(lambda x: x.upper(), df['FirstName']))
print(df)
print("Lower cases in FirstName")
df1['FirstName'] = list(map(lambda x: x.lower(), df['FirstName']))
print(df1)
```
